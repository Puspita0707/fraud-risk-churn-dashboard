# *** *italicized text* Business-Level Insights & Recommendations***

# Customer Insights

- Medium-value customers exhibit the highest churn rate (~27%), indicating a critical segment for retention.
- High-value customers churn less, but still require proactive engagement.
**

#  Fraud Insights

- Fraudulent transactions are typically low-value test charges rather than high-amount transactions.
- Fraud activity is concentrated within specific time windows, suggesting burst-like behavior.
***

#Integrated Business View

- Customer segments with moderate engagement may require closer transaction monitoring, as behavioral patterns associated with fraud resemble those seen in transitional customer segments.
***

# Business Recommendations

- Prioritize retention campaigns for medium-value customers before disengagement escalates.
- Implement real-time alerts for rapid, low-amount transactions occurring in short time intervals.
- Monitor high-risk time windows more closely to detect coordinated fraud activity.
***
